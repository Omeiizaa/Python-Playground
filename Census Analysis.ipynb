<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What is the average age of men?
</div>

```python
df['sex'] = df['sex'].str.strip()
df['age'] = pd.to_numeric(df['age'], errors='coerce')

average_age_men = df[df['sex'] == 'Male']['age'].mean()

print(f"The average age of men is: {average_age_men:.1f} years")
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What is the percentage of people who have a Bachelor's degree?
</div>

```python
df['education'] = df['education'].str.strip()
bachelors_count = df[df['education'] == 'Bachelors'].shape[0]

total_count = df.shape[0]
percentage_bachelors = (bachelors_count / total_count) * 100

# Print the result
print(f"The percentage of people who have a Bachelor's degree is: {percentage_bachelors:.1f}%")
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
</div>

```python
df['over50k'] = df['over50k'].str.strip()
advanced_education = ['Bachelors', 'Masters', 'Doctorate']

advanced_education_df = df[df['education'].isin(advanced_education)]

high_salary_advanced_education = advanced_education_df[advanced_education_df['over50k'] == '>50K']

percentage_high_salary = (high_salary_advanced_education.shape[0] / advanced_education_df.shape[0]) * 100

print(f"The percentage of people with advanced education who make more than 50K is: {percentage_high_salary:.1f}%")
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What percentage of people without advanced education make more than 50K?
</div>

```python
advanced_education = ['Bachelors', 'Masters', 'Doctorate']

non_advanced_education_df = df[~df['education'].isin(advanced_education)]

high_salary_non_advanced_education = non_advanced_education_df[non_advanced_education_df['over50k'] == '>50K']

percentage_high_salary_non_advanced_education = (high_salary_non_advanced_education.shape[0] / non_advanced_education_df.shape[0]) * 100

print(f"The percentage of people without advanced education who make more than 50K is: {percentage_high_salary_non_advanced_education:.1f}%")
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
</div>

```python
# Step 1: Find the minimum number of hours worked per week
min_hours_per_week = df['hoursperweek'].min()

min_hours_people = df[df['hoursperweek'] == min_hours_per_week]

people_with_salary_above_50K = min_hours_people[min_hours_people['over50k'] == '>50K']
percentage_above_50K = (len(people_with_salary_above_50K) / len(min_hours_people)) * 100

print(f"The percentage of people who work the minimum number of hours per week and have a salary of more than 50K is: {percentage_above_50K:.1f}%")
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    What country has the highest percentage of people that earn >50K and what is that percentage?
</div>

```python
country_salary_percentage = df.groupby('nativecountry')['over50k'].value_counts(normalize=True).unstack().fillna(0)

country_salary_percentage['>50K_percentage'] = country_salary_percentage['>50K'] * 100

country_with_highest_percentage = country_salary_percentage['>50K_percentage'].idxmax()
highest_percentage = country_salary_percentage['>50K_percentage'].max()

print(f"The country with the highest percentage of people earning >50K is: {country_with_highest_percentage}")
print(f"The percentage is: {highest_percentage:.1f}%")
```
